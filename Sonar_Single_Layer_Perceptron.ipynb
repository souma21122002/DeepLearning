```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

tf.compat.v1.disable_eager_execution()
```

```python
df=pd.read_csv('/content/drive/MyDrive/ML/sonar.csv',header=None)
print(len(df.columns))
```

```python
df.head()
```

there are two classes 1.rock 2.mine(metal cylinder)

```python
x=df[df.columns[0:60]].values  # x is assigned with 0 to 59th columns which are independent
y=df[df.columns[60]] # y is assigned with the dependent column which is 60th

x
```

```python
y
```

```python
y_labelencoder=LabelEncoder()
y=y_labelencoder.fit_transform(y.astype(str))  # converting the y column from string to numbers
y
```

```python
y=y.reshape(-1,1)
onehotencoder=OneHotEncoder(categories='auto', sparse_output=False)
Y=onehotencoder.fit_transform(y)
Y # in the output if one belongs to rock then its 1 and other is 0 . so we got 2 separate classes
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,Y,test_size=0.20,random_state=42)
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
learning_rate=0.01
training_epochs=100
cost_history=np.empty(shape=[1],dtype=float)
```

```python
n_class=2 #rock and mine
X=tf.compat.v1.placeholder(tf.float32,[None,60]) # 60 columns
w=tf.Variable(tf.zeros([60,n_class])) # 60 size as input are 60s and 2 columns
b=tf.Variable(tf.zeros([n_class])) # 2 class
```

```python
init=tf.compat.v1.global_variables_initializer()
```

```python
y=tf.compat.v1.placeholder(tf.float32,[None,n_class]) #actual y
pred=tf.nn.softmax(tf.matmul(X,w)+b) # softmax is used for multiclass classification as an activation function
```

```python
cost_function=tf.reduce_mean(-tf.reduce_sum(y*tf.math.log(pred))) # cost function calculating rmse
```

```python
training_step=tf.compat.v1.train.GradientDescentOptimizer(learning_rate).minimize(cost_function)
```

```python
sess=tf.compat.v1.Session()
sess.run(init)
mse_history=[]
```

```python
for i in range(100):
  sess.run(training_step,feed_dict={X:train_x,y:train_y})
  cost=sess.run(cost_function,feed_dict={X:train_x,y:train_y})
  cost_history=np.append(cost_history,cost) # storing the cost

  pred_y=sess.run(pred,feed_dict={X:test_x})
  print('epoch:',i,' - ','cost:' , cost)
  mse=tf.reduce_mean(tf.square(pred_y-test_y))
  mse_history.append(sess.run(mse))
```
