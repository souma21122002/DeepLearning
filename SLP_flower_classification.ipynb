```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
tf.compat.v1.disable_eager_execution()
```

```python
flowers_types=os.listdir('/content/drive/MyDrive/ML/Flowers')
flowers_types
```

```python
import cv2
path='/content/drive/MyDrive/ML/Flowers'

im_size=60
images=[]
labels=[]

for i in flowers_types:
  data_path=path+'/'+str(i)
  filenames=[i for i in os.listdir(data_path) if i.endswith('.jpg')]
  for f in filenames:
    img=cv2.imread(data_path+'/'+f)
    img=cv2.resize(img,(im_size,im_size))
    images.append(img)
    labels.append(i)
```

```python
images=np.array(images)
images=images.astype('float32')/255
images
```

```python
images.shape
```

```python
from  sklearn.preprocessing import LabelEncoder , OneHotEncoder
x=images
y=labels
y_labelencoder=LabelEncoder()
y=y_labelencoder.fit_transform(y)
print(y)
```

```python
y=y.reshape(-1,1)
# The 'categorical_features' argument is deprecated.
# Use the 'handle_unknown' and 'sparse' arguments instead.
# Setting 'handle_unknown' to 'ignore' will handle unknown categories during prediction.
# Setting 'sparse' to 'False' will return a dense array instead of a sparse matrix.
onehotencooder=OneHotEncoder(sparse_output=False, handle_unknown='ignore') # Update to use handle_unknown and sparse arguments
Y=onehotencooder.fit_transform(y)
Y
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
X,Y=shuffle(x,Y,random_state=1)
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.20,random_state=415)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
train_x=np.reshape(train_x,(59,(60*60*3)))
print(train_x.shape)

test_x=np.reshape(test_x,(15,(60*60*3)))
print(test_x.shape)
```

```python
learning_rate=0.01
training_epochs=1000
n_dim=60
n_class=2

x=tf.compat.v1.placeholder(tf.float32,[None,10800])
w=tf.Variable(tf.zeros([10800,n_class]))
b=tf.Variable(tf.zeros([n_class]))
y_=tf.compat.v1.placeholder(tf.float32,[None,n_class]) # actual y
```

```python
init=tf.compat.v1.global_variables_initializer()
```

```python
pred=tf.nn.softmax(tf.matmul(x,w)+b)

cost_function=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=pred,labels=y_) )
training_step=tf.compat.v1.train.GradientDescentOptimizer(learning_rate).minimize(cost_function)
```

```python
sess=tf.compat.v1.Session()
sess.run(init)
cost_history=[]
```

```python
for epoch in range(training_epochs):
  sess.run(training_step,feed_dict={x:train_x,y_:train_y})
  cost=sess.run(cost_function,feed_dict={x:train_x,y_:train_y})
  cost_history=np.append(cost_history,cost)
  print('epoch=',epoch,'--','cost=',cost)
pred_y=sess.run(pred,feed_dict={x:test_x})
```
